```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
import seaborn
import random

from sklearn.preprocessing import Normalizer
from sklearn.model_selection import KFold, StratifiedKFold
from sklearn.model_selection import GridSearchCV
from sklearn.svm import LinearSVC
from sklearn.metrics import classification_report, confusion_matrix, f1_score, roc_curve, auc, accuracy_score
from sklearn.cluster import KMeans
from sklearn.cluster import SpectralClustering
from sklearn.neighbors import KNeighborsClassifier

from scipy.special import softmax
from operator import itemgetter

warnings.filterwarnings("ignore")
```

### (a) Download the Breast Cancer Wisconsin (Diagnostic) Data Set

```python
path = '../data/wdbc.data'

cols = ['ID', 'diagnosis']
for i in range(1, 31):
    cols.append('col' + str(i))
```

```python
cancer_df = pd.read_csv(path, header=None)

cancer_df.columns = cols
cancer_df.replace({'M': 1, 'B': 0}, inplace=True)

cancer_df
```

### (b) Monte-Carlo Simulation

```python
def my_train_test_split(df):
    random.seed(100)
    m_indx = df[df['diagnosis'] == 1].index.values
    b_indx = df[df['diagnosis'] == 0].index.values

    m_test_indx = np.random.choice(m_indx, int(len(m_indx)*0.2), replace=False)
    b_test_indx = np.random.choice(b_indx, int(len(b_indx)*0.2), replace=False)

    test_indx = np.concatenate([m_test_indx, b_test_indx], axis=0)

    train_indx = []
    for indx in range(df.shape[0]):
        if indx not in test_indx:
            train_indx.append(indx)
    train_indx = np.array(train_indx)

    train_df = df.iloc[train_indx, :].reset_index(drop=True)
    test_df = df.iloc[test_indx, :].reset_index(drop=True)

    trainX = train_df.iloc[:, 2:]
    trainY = train_df.iloc[:, 1:2]
    testX = test_df.iloc[:, 2:]
    testY = test_df.iloc[:, 1:2]
    
    return trainX, trainY, testX, testY
```

```python
def plot_roc_curve(Y, Y_prob, title):
    
    fig, axes = plt.subplots(figsize=(8, 4))
    axes.set_title("ROC curve for " + str(title))
    
    fpr, tpr, thresholds = roc_curve(Y, Y_prob, pos_label=1)
    roc_auc = auc(fpr, tpr)
    axes.plot(fpr, tpr, label = "ROC Curve Area: " + str(round(roc_auc, 4)))
    axes.plot([0, 1], [0, 1], "--")
    axes.set_xlabel("False Positive Rate")
    axes.set_ylabel("True Positive Rate")
    axes.legend()
    plt.show()
```

#### (i) Supervised Learning

```python
#change 20 for each algorithm
folds = 5

tuning_params_supervised = {'C' : np.logspace(-2, 7, 20)}
classifier_supervised = LinearSVC(penalty = 'l1', dual = False)

normalizer = Normalizer()
```

```python
M = 30
train_metrics = {'Accuracy': [], 'Precision': [], 'Recall': [], 'F1-score': [], 'AUC': [] }
test_metrics = {'Accuracy': [], 'Precision': [], 'Recall': [], 'F1-score': [], 'AUC': [] }

for i in range(M):
    
    #train supervised SVM
    trainX, trainY, testX, testY = my_train_test_split(cancer_df)
    
    trainX_normalized = pd.DataFrame(normalizer.fit_transform(trainX))
    testX_normalized = pd.DataFrame(normalizer.transform(testX))
    trainY = trainY.values.ravel()
    testY = testY.values.ravel()

    cv_supervised = StratifiedKFold(folds, random_state=i, shuffle=True)
    clf = GridSearchCV(estimator = classifier_supervised, param_grid = tuning_params_supervised, cv = cv_supervised, verbose = 0, n_jobs = None)
    clf.fit(trainX_normalized, trainY)

    #----------------------------------------------        
    print("Best parameter setting for iteration " + str(i + 1) + ": " + str(clf.best_params_))
    #----------------------------------------------    
    
    #predict and evaluate for train data
    y_train_pred = clf.predict(trainX_normalized)
    y_train_pred_prob = clf.decision_function(trainX_normalized)
    result = classification_report(trainY, y_train_pred, output_dict=True)

    acc = round(result['accuracy'], 4)
    precision = round(result['weighted avg']['precision'], 4)
    recall = round(result['weighted avg']['recall'], 4)
    f1 = round(result['weighted avg']['f1-score'], 4)
    fpr, tpr, thresholds = roc_curve(trainY, y_train_pred_prob, pos_label = 1)
    auc_score = round(auc(fpr, tpr), 4)

    train_metrics['Accuracy'].append(acc)
    train_metrics['Precision'].append(precision)
    train_metrics['Recall'].append(recall)
    train_metrics['F1-score'].append(f1)
    train_metrics['AUC'].append(auc_score)


    #predict and evaluate for test data
    y_test_pred = clf.predict(testX_normalized)
    y_test_pred_prob = clf.decision_function(testX_normalized)
    test_result = classification_report(testY, y_test_pred, output_dict=True)

    test_acc = round(test_result['accuracy'], 4)
    test_precision = round(test_result['weighted avg']['precision'], 4)
    test_recall = round(test_result['weighted avg']['recall'], 4)
    test_f1 = round(test_result['weighted avg']['f1-score'], 4)
    test_fpr, test_tpr, test_thresholds = roc_curve(testY, y_test_pred_prob, pos_label = 1)
    test_auc_score = round( auc(test_fpr, test_tpr), 4)

    test_metrics['Accuracy'].append(test_acc)
    test_metrics['Precision'].append(test_precision)
    test_metrics['Recall'].append(test_recall)
    test_metrics['F1-score'].append(test_f1)
    test_metrics['AUC'].append(test_auc_score)

    if i == M-1:
        print("\n\nConfusion matrix and ROC for training run")
        train_confusion_matrix = confusion_matrix(trainY, y_train_pred)
        print(train_confusion_matrix)
        plot_roc_curve(trainY, y_train_pred_prob, "training")


        print("Confusion matrix and ROC for testing run")
        test_confusion_matrix = confusion_matrix(testY, y_test_pred)
        print(test_confusion_matrix)
        plot_roc_curve(testY, y_test_pred_prob, "testing")

    
```

#### (ii) Semi-Supervised Learning/ Self-training

##### (A)

```python
tuning_params_semi_supervised = {'C' : np.logspace(-2, 7, 10)}
classifier_semi_supervised = LinearSVC(penalty = 'l1', dual = False)
M = 30
```

##### (B)

```python
train_metrics = {'Accuracy': [], 'Precision': [], 'Recall': [], 'F1-score': [], 'AUC': [] }
test_metrics = {'Accuracy': [], 'Precision': [], 'Recall': [], 'F1-score': [], 'AUC': [] }


for iteration in range(M):
    #prepare labelled and unlabelled datsets
    trainX, trainY, testX, testY = my_train_test_split(cancer_df)
    trainX_normalized = pd.DataFrame(normalizer.fit_transform(trainX))
    testX_normalized = pd.DataFrame(normalizer.transform(testX))

    m_indx = trainY[trainY['diagnosis'] == 1].index.values
    b_indx = trainY[trainY['diagnosis'] == 0].index.values

    m_labelled_indx = np.random.choice(m_indx, int(len(m_indx)*0.5), replace=False)
    b_labelled_indx = np.random.choice(b_indx, int(len(b_indx)*0.5), replace=False)

    labelled_indx = np.concatenate([m_labelled_indx, b_labelled_indx], axis=0)

    unlabelled_indx = []
    for indx in range(trainY.shape[0]):
        if indx not in labelled_indx:
            unlabelled_indx.append(indx)

    unlabelled_indx = np.array(unlabelled_indx)

    labelledX = trainX_normalized.iloc[labelled_indx, :].reset_index(drop=True).to_numpy().tolist()
    labelledY = trainY.iloc[labelled_indx, :].reset_index(drop=True).to_numpy().tolist()
    unlabelledX = trainX_normalized.iloc[unlabelled_indx, :].reset_index(drop=True).to_numpy().tolist()
    #unlabelledY = trainY.iloc[unlabelled_indx, :].reset_index(drop=True).to_numpy().tolist()

    #------------------------------------------
    #train and re-train SVM
    cv_semi_supervised = StratifiedKFold(5, random_state = iteration, shuffle=True)
    semi_supervised_clf = GridSearchCV(estimator = classifier_semi_supervised, param_grid = tuning_params_semi_supervised, cv = cv_semi_supervised, verbose = 0, n_jobs = None)
    semi_supervised_clf.fit(labelledX, labelledY)

    while len(unlabelledX) > 0:

        #decision_function directly proportional to distance from hyperplane.
        scores = semi_supervised_clf.decision_function(unlabelledX)

        #find farthest point index in unlabelled data X
        max_distance = float('-inf')
        for i in range(len(scores)):
            if max_distance < scores[i]:
                max_distance = scores[i]
                indx = i


        #predict its label
        x = unlabelledX.pop(indx)
        y = semi_supervised_clf.predict(np.array([x]))

        labelledX.append(x)
        labelledY.append(y)

        #re-train the SVM
        semi_supervised_clf.best_estimator_.fit(labelledX, labelledY)

    #----------------------------------------------        
    print("Best parameter setting for iteration " + str(iteration + 1) + ": " + str(semi_supervised_clf.best_params_))
    #----------------------------------------------    


    #predict and evaluate for train data
    y_train_pred = semi_supervised_clf.predict(trainX_normalized)
    y_train_pred_prob = semi_supervised_clf.decision_function(trainX_normalized)
    result = classification_report(trainY, y_train_pred, output_dict=True)

    acc = round(result['accuracy'], 4)
    precision = round(result['weighted avg']['precision'], 4)
    recall = round(result['weighted avg']['recall'], 4)
    f1 = round(result['weighted avg']['f1-score'], 4)
    fpr, tpr, thresholds = roc_curve(trainY, y_train_pred_prob, pos_label = 1)
    auc_score = round(auc(fpr, tpr), 4)

    train_metrics['Accuracy'].append(acc)
    train_metrics['Precision'].append(precision)
    train_metrics['Recall'].append(recall)
    train_metrics['F1-score'].append(f1)
    train_metrics['AUC'].append(auc_score)


    #predict and evaluate for test data
    y_test_pred = semi_supervised_clf.predict(testX_normalized)
    y_test_pred_prob = semi_supervised_clf.decision_function(testX_normalized)
    test_result = classification_report(testY, y_test_pred, output_dict=True)

    test_acc = round(test_result['accuracy'], 4)
    test_precision = round(test_result['weighted avg']['precision'], 4)
    test_recall = round(test_result['weighted avg']['recall'], 4)
    test_f1 = round(test_result['weighted avg']['f1-score'], 4)
    test_fpr, test_tpr, test_thresholds = roc_curve(testY, y_test_pred_prob, pos_label = 1)
    test_auc_score = round( auc(test_fpr, test_tpr), 4)

    test_metrics['Accuracy'].append(test_acc)
    test_metrics['Precision'].append(test_precision)
    test_metrics['Recall'].append(test_recall)
    test_metrics['F1-score'].append(test_f1)
    test_metrics['AUC'].append(test_auc_score)

    if iteration == M-1:
        print("\n\nConfusion matrix and ROC for training run")
        train_confusion_matrix = confusion_matrix(trainY, y_train_pred)
        print(train_confusion_matrix)
        plot_roc_curve(trainY, y_train_pred_prob, "training")


        print("Confusion matrix and ROC for testing run")
        test_confusion_matrix = confusion_matrix(testY, y_test_pred)
        print(test_confusion_matrix)
        plot_roc_curve(testY, y_test_pred_prob, "testing")
```

#### (iii) Unsupervised Learning

##### (A)

Setting the number of iterations allows us to avoid getting stuck in local minima trap for k-means.

##### (B)

```python
def get_distance_from_centers(centroids, X, predicted_lables):
    dist_arr_0 = []
    dist_arr_1 = []
    
    prob_arr = np.ndarray((X.shape[0], 2))

    #find distance of every point to both cluster centroids
    for i in range(len(X)):
    
        sample_point_X = X[i]

        dist_arr_0.append( np.linalg.norm(sample_point_X - centroids[0]))
        dist_arr_1.append( np.linalg.norm(sample_point_X - centroids[1]))

        prob_arr[i, 0] = np.linalg.norm(sample_point_X - centroids[0])
        prob_arr[i, 1] = np.linalg.norm(sample_point_X - centroids[1])

    
    prob_arr = softmax(prob_arr, axis=1).astype(np.float64)
    
    return dist_arr_0, dist_arr_1, prob_arr


def majority_poll(dist_arr_0, dist_arr_1, predicted_labels, Y):
    
    cluster_major_labels = {}
    
    indices_0 = list(Y.index.values)
    closest_30_tocluster_0_indx = sorted(indices_0, key = lambda i:dist_arr_0[i])[:30]
    majority_label = Y.iloc[closest_30_tocluster_0_indx].value_counts().index[0]
    cluster_major_labels['cluster'+str(0)] = majority_label[0]
    
    indices_1 = list(Y.index.values)
    closest_30_tocluster_1_indx = sorted(indices_1, key = lambda i:dist_arr_1[i])[:30]
    majority_label = Y.iloc[closest_30_tocluster_1_indx].value_counts().index[0]
    cluster_major_labels['cluster'+str(1)] = majority_label[0]
    
    return cluster_major_labels
```

```python
train_metrics = {'Accuracy': [], 'Precision': [], 'Recall': [], 'F1-score': [], 'AUC': [] }
test_metrics = {'Accuracy': [], 'Precision': [], 'Recall': [], 'F1-score': [], 'AUC': [] }

M = 30
for iteration in range(M):
    trainX, trainY, testX, testY = my_train_test_split(cancer_df)

    trainX_normalized = normalizer.fit_transform(trainX)
    testX_normalized = normalizer.transform(testX)

    #set random_state to iteration value
    kmeans_model = KMeans(n_clusters = 2, n_init = 15, random_state = iteration).fit(trainX_normalized)

    centroids = kmeans_model.cluster_centers_
    predicted_labels = kmeans_model.predict(trainX_normalized)
    test_predicted_labels = kmeans_model.predict(testX_normalized)

    #train
    dist_0, dist_1, y_pred_prob_train = get_distance_from_centers(centroids, trainX_normalized, predicted_labels)

    cluster_major_labels = majority_poll(dist_0, dist_1, predicted_labels, trainY)

    indx_0 = np.where(predicted_labels == 0)
    indx_1 = np.where(predicted_labels == 1)
    predicted_labels[indx_0] = cluster_major_labels['cluster0']
    predicted_labels[indx_1] = cluster_major_labels['cluster1']


    #train metrics
    positive_cluster = cluster_major_labels['cluster0']
    y_train_pred = list(predicted_labels.tolist())
    y_train_pred_prob = y_pred_prob_train[:, positive_cluster]

    result = classification_report(trainY, y_train_pred, output_dict=True)

    acc = round(result['accuracy'], 4)
    precision = round(result['weighted avg']['precision'], 4)
    recall = round(result['weighted avg']['recall'], 4)
    f1 = round(result['weighted avg']['f1-score'], 4)
    fpr, tpr, thresholds = roc_curve(trainY, y_train_pred_prob, pos_label = 1)
    auc_score = round(auc(fpr, tpr), 4)

    train_metrics['Accuracy'].append(acc)
    train_metrics['Precision'].append(precision)
    train_metrics['Recall'].append(recall)
    train_metrics['F1-score'].append(f1)
    train_metrics['AUC'].append(auc_score)


    #test
    test_dist_0, test_dist_1, y_pred_prob_test = get_distance_from_centers(centroids, testX_normalized, test_predicted_labels)

    test_cluster_major_labels = majority_poll(test_dist_0, test_dist_1, test_predicted_labels, testY)

    test_indx_0 = np.where(test_predicted_labels == 0)
    test_indx_1 = np.where(test_predicted_labels == 1)
    test_predicted_labels[test_indx_0] = test_cluster_major_labels['cluster0']
    test_predicted_labels[test_indx_1] = test_cluster_major_labels['cluster1']


    #test metrics
    test_positive_cluster = test_cluster_major_labels['cluster0']
    y_test_pred = list(test_predicted_labels.tolist())
    y_test_pred_prob = y_pred_prob_test[:, test_positive_cluster]

    test_result = classification_report(testY, y_test_pred, output_dict=True)

    acc = round(test_result['accuracy'], 4)
    precision = round(test_result['weighted avg']['precision'], 4)
    recall = round(test_result['weighted avg']['recall'], 4)
    f1 = round(test_result['weighted avg']['f1-score'], 4)
    fpr, tpr, thresholds = roc_curve(testY, y_test_pred_prob, pos_label = 1)
    auc_score = round(auc(fpr, tpr), 4)

    test_metrics['Accuracy'].append(acc)
    test_metrics['Precision'].append(precision)
    test_metrics['Recall'].append(recall)
    test_metrics['F1-score'].append(f1)
    test_metrics['AUC'].append(auc_score)
    
    print("Iteration " + str(iteration+1))
    
    if iteration == M-1:
        print("\n\nConfusion matrix and ROC for training run")
        train_confusion_matrix = confusion_matrix(trainY, y_train_pred)
        print(train_confusion_matrix)
        plot_roc_curve(trainY, y_train_pred_prob, "training")


        print("Confusion matrix and ROC for testing run")
        test_confusion_matrix = confusion_matrix(testY, y_test_pred)
        print(test_confusion_matrix)
        plot_roc_curve(testY, y_test_pred_prob, "testing")
```

```python
M = 50
passive_test_error = []

for iteration in range(M):
    
    train_df, test_df = random_train_test_split(banknote_df)
    trainX, trainY, testX, testY = get_x_y_split(train_df, test_df)
    trainY = trainY.reset_index(drop=True)
    testY = testY.reset_index(drop = True)

    passive_cv = StratifiedKFold(passive_folds, random_state = iteration, shuffle = True)

    test_error_arr = []
    train_indx = []
    remaining_indx = np.arange(trainX.shape[0])

    while len(remaining_indx) > 0:
        
        np.random.shuffle(remaining_indx)

        pool_indx = remaining_indx[:10]
        train_indx.extend(pool_indx)
        remaining_indx = remaining_indx[10:]

        passive_trainX = trainX.iloc[train_indx]
        passive_trainY = trainY.iloc[train_indx]

        passive_clf = GridSearchCV(estimator = classifier,
                                   param_grid = passive_tuning_params,
                                   cv = passive_cv,
                                   verbose = 0,
                                   n_jobs = None)

        passive_clf.fit(passive_trainX, passive_trainY)

        y_pred = passive_clf.predict(testX)

        test_accuracy = classification_report(testY, y_pred, output_dict = True)['accuracy']
        test_error = round(1 - test_accuracy, 4)
        test_error_arr.append(test_error)
        
    avg_test_error = sum(test_error_arr) / len(test_error_arr)
    print("Iteration " + str(iteration+1) + ": Average test error = " + str(round(avg_test_error, 4)))
    
    passive_test_error.append(test_error_arr)
```

#### (ii) Active

```python
M = 50

active_test_error = []
for iteration in range(M):

    train_df, test_df = random_train_test_split(banknote_df)
    trainX, trainY, testX, testY = get_x_y_split(train_df, test_df)
    trainY = trainY.reset_index(drop=True)
    testY = testY.reset_index(drop = True)
    
    active_cv = StratifiedKFold(5, random_state = iteration, shuffle = True)

    test_error_arr = []
    train_indx = []
    remaining_indx = np.arange(trainX.shape[0])

    np.random.shuffle(remaining_indx)
    pool, remaining_indx = remaining_indx[:10], remaining_indx[10:]
    train_indx.extend(pool)

    remaining_df = trainX.iloc[remaining_indx]
    
    
    active_trainX, active_trainY = pd.DataFrame(), pd.DataFrame()


    while len(remaining_indx) >= 0:
        
        active_trainX, active_trainY = trainX.iloc[train_indx], trainY.iloc[train_indx]
    
        
        active_clf = GridSearchCV(estimator = classifier,
                           param_grid = passive_tuning_params,
                           cv = active_cv,
                           verbose = 0,
                           n_jobs = None)
        active_clf.fit(active_trainX, active_trainY)
        y_pred = active_clf.predict(testX)
        result = classification_report(testY, y_pred, output_dict = True)
        test_error = round(1 - result['accuracy'], 4)
        test_error_arr.append(test_error)


        if len(remaining_indx) == 0:
            active_test_error.append(test_error_arr)
            
            avg_test_error = sum(test_error_arr) / len(test_error_arr)
            print("Iteration " + str(iteration+1) + ": Average test error = " + str(round(avg_test_error, 4)))
    
            break

        dist = active_clf.decision_function(remaining_df)
        D = {}
        pool = []

        for i in range(len(dist)):
            D[remaining_df.index[i]] = dist[i]
        dist1 = sorted(D.items(), key = itemgetter(1))[:10]
        
        for i in range(10):
            pool.append(dist1[i][0])


        remaining_indx = np.setdiff1d(remaining_indx, pool)
        remaining_df = remaining_df.loc[remaining_indx]
        train_indx.extend(pool)
        
```
